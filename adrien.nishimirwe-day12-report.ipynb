Exercise 1:
Create a simple line plot for $y = x$ for x values ranging from 0 to 10.

Exercise 9:
Create a subplot with 2 rows and 1 column. Plot $y_{1} = x^2$ in the first subplot and $y_{2} = x^3$ in the second.

```python
x = np.linspace(-10, 10, 100)  

y1 = x**2   
y2 = x**3  
fig, axs = plt.subplots(2, 1, figsize=(8, 10)) 

# Plot y1 = x^2 in the first subplot  
axs[0].plot(x, y1, color='blue')  
axs[0].set_title('y = x^2')  
axs[0].set_xlabel('x')  
axs[0].set_ylabel('y')  
axs[0].grid(True)  

# Plot y2 = x^3 in the second subplot  
axs[1].plot(x, y2, color='orange')  
axs[1].set_title('y = x^3')  
axs[1].set_xlabel('x')  
axs[1].set_ylabel('y')  
axs[1].grid(True)  
plt.tight_layout()  
plt.show()
```

Exercise 21:
Plot a contour plot for the function $z = sin(x) + cos(y)$ over a grid of x and y values ranging from -5 to 5.

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = np.sin(X) + np.cos(Y)

plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')

plt.colorbar(contour)
plt.title(r'Contour plot of $z = \sin(x) + \cos(y)$')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

Exercise 25:
Create a plot with logarithmic scaling on both axes.

```python
x = np.logspace(0, 3, 100)  
y = np.sqrt(x)              

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = sqrt(x)')

plt.xscale('log')
plt.yscale('log')

plt.title('Log-Log Plot of y = sqrt(x)')
plt.xlabel('x')
plt.ylabel('y')

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

plt.show()
```

Exercise 26:
Create a 3D plot of the function $z = x^2 + y^2$ over a grid of x and y values.

```python
x = np.linspace(-5, 5, 100) 
y = np.linspace(-5, 5, 100)  
X, Y = np.meshgrid(x, y)      
Z = X**2 + Y**2             

fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_title('3D Plot of z = x^2 + y^2')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

Exercise 27:
Create a radar chart with 5 variables and random values between 0 and 1.

```python
num = 5
angles = np.linspace(0, 2 * np.pi, num, endpoint=False).tolist()
values = np.random.rand(num).tolist()

values += values[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='blue', alpha=0.25) 
ax.plot(angles, values, color='blue', linewidth=2)  

labels = [f'Variable {i+1}' for i in range(num)]
plt.xticks(angles[:-1], labels) 

plt.title('Radar Chart with 5 Variables')

plt.show()
```
